## 1. Configuration

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

## 2. Data

```python
auto = pd.read_csv('auto-mpg.csv')
auto
```

## 3. Visualizations

### a) Default & Circle & Point

```python
# # Use these settings to enlarge figure & font size
# plt.rcParams['figure.figsize'] = (16.0, 8.0)
# plt.rcParams.update({'font.size': 16})

auto.plot(kind="scatter", x="hp", y="mpg", title="Default marker")
auto.plot(kind="scatter", x="hp", y="mpg", title="Circle", marker="o", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Point", marker=".", s=100)
```

### b) Line

```python
auto.plot(kind="scatter", x="hp", y="mpg", title="Vertical line", marker="|", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Horizontal line", marker="_", s=100)
```

### c) Plus

```python
auto.plot(kind="scatter", x="hp", y="mpg", title="Plus", marker="+", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Plus (filled)", marker="P", s=100)
```

### d) X

```python
auto.plot(kind="scatter", x="hp", y="mpg", title="x", marker="x", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="x (filled)", marker="X", s=100)
```

### e) Triangle

```python
auto.plot(kind="scatter", x="hp", y="mpg", title="Triangle down", marker="v", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Triangle up", marker="^", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Triangle left", marker="<", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Triangle right", marker=">", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Tri down", marker="1", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Tri up", marker="2", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Tri left", marker="3", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Tri right", marker="4", s=100)
```

### f) Pixel & Square

```python
auto.plot(kind="scatter", x="hp", y="mpg", title="Pixel", marker=",", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Square", marker="s", s=100)
```

### g) Diamond

```python
# Diamond
auto.plot(kind="scatter", x="hp", y="mpg", title="Diamond", marker="D", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Thin diamond", marker="d", s=100)
```

### h) Star

```python
auto.plot(kind="scatter", x="hp", y="mpg", title="Star", marker="*", s=100)
```

### i) Pentagon & Hexagon & Octagon

```python
# Pentagon
auto.plot(kind="scatter", x="hp", y="mpg", title="Pentagon", marker="p", s=100)

# Hexagon
auto.plot(kind="scatter", x="hp", y="mpg", title="Hexagon1", marker="h", s=100)
auto.plot(kind="scatter", x="hp", y="mpg", title="Hexagon2", marker="H", s=100)

# Octagon
auto.plot(kind="scatter", x="hp", y="mpg", title="Octagon", marker="8", s=400)
```
